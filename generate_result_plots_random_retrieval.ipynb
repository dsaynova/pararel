```python
data = pd.read_csv("data/results/2023_06_14_retrieval_statistics_random.csv")
data.head()
```

```python
data.columns
```

```python
def get_mean_std_string(row):
    return f"${row.loc['mean']:.2f} \pm {row.loc['std']:.2f}$"
```

```python
cols_of_interest = ["retriever_id_consistency", "retriever_id_consistency_std"]
tmp_data = data.agg({col: ["mean", "std"] for col in cols_of_interest})
cols = tmp_data.columns.get_level_values(0).unique()
#data_to_concat = (tmp_data[col].apply(get_mean_std_string, axis=1) for col in cols)
print_data = tmp_data.apply(get_mean_std_string, axis=0)

#print_data = pd.concat(data_to_concat, axis=1, keys=cols)
print(print_data.to_latex(escape=False))
```

```python
cols_of_interest = ["retriever_id_consistency", "retriever_id_consistency_std", "retriever_title_consistency", "retriever_title_consistency_std", "retriever_embedding_similarity_consistency", "retriever_embedding_similarity_consistency_std"]
tmp_data = data.groupby(["sample_across"]).agg({col: ["mean", "std"] for col in cols_of_interest})
cols = tmp_data.columns.get_level_values(0).unique()
data_to_concat = (tmp_data[col].apply(get_mean_std_string, axis=1) for col in cols)
print_data = pd.concat(data_to_concat, axis=1, keys=cols)
print(print_data.to_latex(escape=False))
```

```python
corr_cols = {'corr_retriever_emb_id_consistency': ('embedding', 'id'),
            'corr_retriever_emb_title_consistency': ('embedding', 'title'),
            'corr_retriever_id_title_consistency': ('id', 'title')}
cols_of_interest = corr_cols.keys()
tmp_data = data.fillna(0).groupby(["sample_across"]).agg({col: ["mean", "std"] for col in cols_of_interest})

unique_cols = ['id', 'title', 'embedding']
corr_data = pd.DataFrame(np.ones((len(unique_cols), len(unique_cols))), columns=unique_cols, index=unique_cols)
for col, vals in corr_cols.items():
    corr_data.at[vals[0], vals[1]] = get_mean_std_string(tmp_data[col].loc['all'])
    corr_data.at[vals[1], vals[0]] = get_mean_std_string(tmp_data[col].loc['all'])
# fix diagonal
for col in unique_cols:
    corr_data.at[col, col] = "$1.00 \pm 0.00$"
corr_data
```
